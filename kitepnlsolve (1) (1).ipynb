SEPERATED QTY COLUMN INTO PLACED AND EXECUTED QUANTITY

```python
df[['executed_qty', 'placed_qty']] = df['Qty.'].str.split('/', expand=True)
df['executed_qty'] = df['executed_qty'].astype(int)
df['placed_qty'] = df['placed_qty'].astype(int)
```

MIS = INTRADAY ONLY
EXECUTED QTY SHOULD BE GREATER THAN 0
STATUS SHOULD BE COMPLETE

```python
df = df[df['Product'] ==  'MIS']
df = df[df['executed_qty'] > 0]
df = df[df['Status'] == 'COMPLETE']
```

```python
df
```

TURNOVER = QUANTITY * AVG PRICE

```python
df['turnover'] = df['executed_qty'] * df['Avg. price']
```

```python
df.head()
```

BROKERAGE CHARGES

```python
def brokerage (turnover):
    return min(20, 0.0003 * turnover)
```

STT - SECURITIES TRANSACTION CHARGES 

```python
def stt(turnover, trade_type):
    return turnover * 0.00025 if trade_type == 'SELL' else 0 
```

ETC - EXCHANGE TRANSACTION CHARGES

```python
def exchange_txn(turnover):
    return turnover * 0.0000325
```

SEBI CHARGES 

```python
def sebi_charges(turnover):
    return turnover * 0.000001
```

GST CALCULATION 

```python
def gst(brokerage, exchange): 
    return 0.18 * (brokerage + exchange)
```

STAMP DUTY

```python
def stamp_duty(turnover): 
    return turnover * 0.00003
```

```python
df['Brokerage'] = df['turnover'].apply(brokerage)
```

```python
df['Brokerage']
```

```python
df.head()
```

```python
df['STT'] = df.apply( lambda x: stt(x['turnover'], x['Type']), axis=1)
```

```python
df['STT']
```

```python
df['Exchange'] = df['turnover'].apply(exchange_txn)
```

```python
df['SEBI'] = df['turnover'].apply(sebi_charges)
```

```python
df['Stamp'] = df['turnover'].apply(stamp_duty)
```

```python
df.head()
```

```python
df['GST'] = df.apply(lambda x: gst(x['Brokerage'], x["Exchange"]), axis=1)
```

TOTAL CHARGES = BROKERAGE + STT + EXCHANGE + SEBI + GST + STAMP 

```python
df["Total_charges"] = df[
    ['Brokerage', 'STT', 'Exchange','SEBI', 'GST', 'Stamp']
].sum(axis=1)
```

solution 2 

```python
solu2 = df.groupby(['Instrument', 'Type']).agg(
    Qty = ('executed_qty', 'sum'),
    Turnover = ('turnover', 'sum'),
    Brokerage = ('Brokerage', 'sum'), 
    STT_CTT = ('STT', 'sum'),
    ETC = ('Exchange', 'sum'), 
    SEBI = ('SEBI','sum'), 
    GST = ('GST', 'sum'), 
    Stamp_charges = ('Stamp', 'sum'), 
    Total_charges = ('Total_charges', 'sum')
).reset_index()
```

WEIGHTED AVERAGE

```python
solu2['weighted_avg'] = solu2['Turnover'] / solu2['Qty']
```

```python
solu2 = solu2[
    [
        'Instrument', 'Type', 'Qty', 'weighted_avg', 'Turnover', 'Brokerage', 'STT_CTT', 'ETC', 'SEBI', 'GST', 'Stamp_charges', 'Total_charges'
    ]
]
```

```python
buy = df[df['Type'] == 'Buy'].groupby('Instrument')['turnover'].sum()
sell = df[df['Type'] == 'Sell'].groupby('Instrument')['turnover'].sum()
charges = df.groupby('Instrument')['Total_charges'].sum()

summary = pd.DataFrame({
    'Buy Value': buy,
    'Sell Value': sell,
    'Total_charges': charges
}).fillna(0)

summary['Gross PnL'] = summary['Sell Value'] - summary['Buy Value']
summary['Net PnL'] = summary['Gross PnL'] - summary['Total_charges']

summary['% Charges on Gross PnL'] = summary.apply(
    lambda x: (x['Total Charges'] / abs(x['Gross PnL']) * 100)
    if x['Gross PnL'] != 0 else 0,
    axis=1
)

summary = summary[['Gross PnL','Total_charges','Net PnL','% Charges on Gross PnL']]
```
